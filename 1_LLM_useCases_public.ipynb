````python
# This notebook does not contain the OpenAI API key!!!

client = OpenAI(
    api_key = ''
)

def chat_with_gpt(prompt, role="user", temperature=0):

  messages=[{"role": role, "content": prompt}]

  response = client.chat.completions.create(
    model="gpt-4-turbo",
    messages=messages,
    temperature=temperature,
    max_tokens=1024
  )

  result = response.choices[0].message.content
  return result

prompt = f"""
What is the capital of Hungary?
"""
result = chat_with_gpt(prompt)
print(result)
````

````python
from redlines import Redlines
from IPython.display import display, HTML

text = [
  "The girl with the black and white puppies have a ball.",  # The girl has a ball.
  "Yolanda has her notebook.", # ok
  "Its going to be a long day. Does the car need it’s oil changed?",  # Homonyms
  "Their goes my freedom. There going to bring they’re suitcases.",  # Homonyms
  "Your going to need you’re notebook.",  # Homonyms
  "That medicine effects my ability to sleep. Have you heard of the butterfly affect?", # Homonyms
  "This phrase is to cherck chatGPT for speling abilitty"  # spelling
]
for t in text:
    prompt = f"""Proofread and correct the following text
    and rewrite the corrected version. If you don't find
    any errors, just say "No errors found". Don't use
    any punctuation around the text:
    ```{t}```"""

    response = chat_with_gpt(prompt)

    if response=="No errors found" or response=="No errors found.":
      print(t,": ",response)
    else:
      diff = Redlines(t,response)
      display(HTML(diff.output_markdown))
````

````python
import pandas as pd
import matplotlib.pyplot as plt

f = open("launches.json")
data = json.load(f)

# Convert data to DataFrame
df = pd.DataFrame(data)

# Convert 'launch_year' to integer
df['launch_year'] = df['launch_year'].astype(int)

# Group by 'launch_year' and 'launch_success' and count occurrences
launches_per_year = df.groupby(['launch_year', 'launch_success']).size().unstack(fill_value=0)

# Plot data
launches_per_year.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Number of Successful and Unsuccessful Launches per Year')
plt.xlabel('Year')
plt.ylabel('Number of Launches')
plt.legend(title='Launch Success', labels=['Unsuccessful', 'Successful'])
plt.show()
````
